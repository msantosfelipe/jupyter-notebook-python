**Múltiplos plots**

```python
plt.plot(array, seno) #azul
plt.plot(array, -1*seno)
```

**Tamanho do gráfico**

```python
figura = plt.figure(figsize=(20,10)) # largura, altura
plt.plot(array, seno)
plt.plot(array, -1*seno)
```

**Titulo e nomes dos eixos do gráfico**

```python
plt.plot(array, seno)
plt.plot(array, -1*seno)
plt.title('Função seno(x) e (-1)Seno(x)')
plt.xlabel('X')
plt.ylabel('Seno(x)')
```

**Personalização do gráfico**

```python
## Mudando de cor e colocando pontos e diferentes
figure = plt.figure(figsize=(15,20))
plt.title('Função Seno e coseno variadas', fontsize=20)
plt.plot(array, seno, color='green', marker='o') #usando argumentos explicitamente
plt.plot(array, -1*seno, 'r-^') #usando string de formatação
plt.plot(array, np.cos(array), 'b-1', markersize=15, markeredgewidth=4) # tamanho e espessura
plt.plot(array, -1*np.cos(array), 'y-', linewidth=5) # espessura da linha
```

```python
# subplots
figure = plt.figure(figsize=(15,10))

# ('linha', 'coluna', 'celula')
plt.subplot(2,3,1) # primeiro
plt.plot(seno, array, 'r*-')

plt.subplot(2,3,2) # segundo
plt.plot(array, seno, 'g--')

plt.subplot(2,3,3) # terceiro
plt.plot(array, array, linewidth=5, linestyle='-')

plt.subplot(2,3,4) # quarto
plt.plot(-1*array, array, 'g*')
```

```python
plt.plot(array, seno, label='Seno (x)')
plt.xlabel('X')
plt.ylabel('Seno(x)')
#plt.legend() # legenda default
#plt.legend(frameon=False, loc='upper right')
plt.legend(bbox_to_anchor=[1,1])
plt.grid(True)
```

### Outros estilos

```python
# criar alguns dados
x1 = np.array([1,2,3,4,5])
y1 = x1*2

x2 = np.linspace(0,10,50)
y2 = x2 * np.random.rand(x2.size)*10
```

```python
# definindo espaço do gráfico
fig, axes = plt.subplots(1,4,figsize=[15,3])

axes[0].scatter(x2,y2, color='green')
axes[0].set_title('Scatter plot')

axes[1].step(x1,y1, color='red')
axes[1].set_title('Step plot')

axes[2].bar(x2,y2, color='yellow')
axes[2].set_title('Bar plot')

axes[3].fill_between(x1,y1*1.0, y1*0.1, color='green', alpha=0.3)
axes[3].plot(x1,y1, color='black')
axes[3].set_title('Linhas preenchidas')
```
