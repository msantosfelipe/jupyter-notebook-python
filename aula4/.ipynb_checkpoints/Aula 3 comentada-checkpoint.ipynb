## Matplotlib

**Importando módulos necessários:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#linha mégica para  funcionar no Jupyter
%matplotlib inline
```

**Criando dados**

```python
array = np.linspace(0, 10, 100) #criando
```

```python
array #inspecionando
```

**Gráfico simples**

```python
plt.plot(array);
#plt.show() #se não estivermos no jupyter!!
```

**Gráfico da função Seno**

```python
seno = np.sin(array) #criando
```

```python
seno #inspecionando
```

```python
plt.plot(array, seno);
```

**Exportando gráfico**

```python
figura = plt.figure() #criando o espaço gráfico
plt.plot(array, seno); #plotando
```

```python
figura.savefig('teste.png') #salvando a imagem
```

**Múltiplos `plots`**

```python
plt.plot(array, seno) #primeira linha - azul
plt.plot(array, -1*seno) #segunda linha
```

**Tamanho do gráfico**

```python
figura = plt.figure(figsize=(20,10)) #REFERÊNCIA: (largura do plot, altura do plot)
plt.plot(array, seno)
plt.plot(array, -1*seno)
```

**Título e nomes dos eixos do gráfico**

```python
plt.plot(array, seno)
plt.plot(array, -1*seno)
plt.title('Função Seno(x) e (-1)Seno(x)')
plt.xlabel('X')
plt.ylabel('Seno(x)');
```

**Personalizações do gráfico**

```python
#Mudando de cor e colocando pontos e diferentes
figure = plt.figure(figsize=(15, 10))
plt.title('Função Seno e Cosseno variadas', fontsize=20)
plt.plot(array, seno, color='green', marker='o') #usando os argumentos explicitamente
plt.plot(array, -1*seno, 'r-^') #usando string de formatação
plt.plot(array, np.cos(array), 'b-1', markersize=15, markeredgewidth=4) #tamanho e espessura
plt.plot(array, -1*np.cos(array), 'y-', linewidth=5) #espessura da linha
```

```python
#subplots
figure = plt.figure(figsize=(15,10)) #ajustando o tamanho

## REFERÊNCIA: ("linha", "coluna", "célula")
plt.subplot(2, 3, 1) #primeiro
plt.plot(seno, array, 'r*-')
plt.title('Primeiro')

plt.subplot(2, 3, 2) #segundo
plt.plot(array, seno, 'g--')
plt.title('Segundo')

plt.subplot(2, 3, 3) #terceiro
plt.plot(array, array, linewidth=5, linestyle='-')
plt.title('Terceiro')

plt.subplot(2, 3, 4) #quarto
plt.plot(-1*array, array, 'g*')
plt.title('Quarto');
```

```python
#definindo intervalo de valores para os eixos
plt.plot(array, seno)

#definindo intervalos
plt.xlim([0, 3.2]) #"recortando" meu eixo X
plt.ylim([0, 1.1]) #"recortando" meu eixo Y
```

```python
#adicionar legendas
plt.plot(array, seno, label='Seno(X)') #incluindo o argumento "label"

plt.xlabel('X') #nome do eixo X
plt.ylabel('Seno(x)') #nome do eixo y

#plt.legend() #criando a legenda - opcao default
#plt.legend(frameon=False, loc='upper right') #criando a legenda - opcao personalizada
plt.legend(bbox_to_anchor=(1,1)) #criando a legenda - opcao personalizada
plt.grid(True) #cria uma grade na figura
```

## Outros estilos

```python
#criar alguns dados
x1 = np.array([1, 2, 3, 4, 5])
y1 = x1**2

x2 = np.linspace(0, 10, 50)
y2 = x2 + np.random.rand(x2.size)*10
```

```python
#definindo espaço de gráficos
fig, axes = plt.subplots(1, 4, figsize=(15,3))

#primeiro gráfico
axes[0].scatter(x2, y2, color='green')
axes[0].set_title('Scatter plot')

#segundo gráfico
axes[1].step(x1, y1, color='red')
axes[1].set_title('Step plot')

#terceiro gráfico
axes[2].bar(x1, y1, color='yellow')
axes[2].set_title('Bar plot')

#quarta gráfico
axes[3].fill_between(x1, y1*1.9, y1*0.1, color='green', alpha=0.3)
axes[3].plot(x1, y1, color='black')
axes[3].set_title('Linhas preenchidas')
```

```python
#gráfico de pizza
estados = ['SP', 'MG', 'RJ', 'ES']
pops = [44.04, 20.87, 16.72, 3.8]
```

```python
plt.pie(pops, labels=estados, autopct='%.2f%%', explode=[0, 0.14, 0, 0.4]);
```

```python
#gráfico de contorno

#criando a funcao
def f(x, y):
    return (1 - x/2 + x**5 + y**3)*np.exp(-x**2 - y**2)

#criando variaveis
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)
```

```python
#criando p grafico de superficie
plt.contourf(X, Y, Z, 18, alpha=0.75, cmap='jet')
```

```python
#pegando somente o contorno
plt.contour(X,Y,Z, colors='black')
```

```python
plt.contourf(X, Y, Z, alpha=0.75, cmap='jet')
plt.contour(X,Y,Z, colors='black')
```

```python
#superficie 3D
from mpl_toolkits.mplot3d import Axes3D
```

```python
#criando espaço de gráficos
fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z, cmap='jet')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
#rotacionando o gráfico
ax.view_init(59, 140)
fig
```

## Seaborn

**Importando módulos**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Pegando dados do `seaborn`**

```python
sns.get_dataset_names() #checando conjuntos de dados disponíveis "dentro" do próprio `seaborn`
```

```python
#importando dados
tips = sns.load_dataset('tips')
```

```python
tips.head() #inpecionando
```

```python
tips.info() #inspecionando
```

```python
#bar plot: que dia gasta mais
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
#grafico gorjeta X dia
sns.barplot(x='tip', y='day', data=tips, orient='h') #orient = parametro para orientacao do gráfico
```

```python
#graficos pareados
sns.barplot(x='day', y='tip', hue='sex', data=tips)
```

```python
#grafico de contagem
sns.countplot(x='sex', data=tips)
```

```python
#diagrama de caixa
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
#combinando informações
plt.figure(figsize=(7,7))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

```python
#gráfico de violino
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
```

```python
#verificando distribuição
sns.distplot(tips['total_bill'])
```

```python
#verificando distribuição
sns.distplot(tips['total_bill'], bins=50) #bins = número de intervalos no histograma
```

```python
#retirando kde o colocando rug
sns.distplot(tips['tip'], kde=False, rug=True)
```

```python
#plotando apenas o kde
sns.kdeplot(tips['total_bill'])
```

```python
#plotando um scatter plot (pontos) com reta de tendência
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total da conta')
plt.ylabel('Gorjeta')
```

```python
#plotando um scatter plot (pontos) com polinômio de grau 3 (por exemplo)
sns.lmplot(x='total_bill', y='tip', data=tips, order=3)
plt.text(10, 0, 'equacao')
```

```python
#plotando sem a linha de tendência
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
#enriquecendo o gráfico (histograma nas laterais)
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
##enriquecendo o gráfico (hexagonos com cores de acordo com a densidade de pontos)
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
#verificando relação entre as colunas do DataFrame
sns.pairplot(tips)
```

```python
#testando para os dados iris
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris);
```

```python
#calculando correlações no conjunto de dados
iris.corr()
```

```python
corr = iris.corr() #salvando correlação em um objeto
```

```python
sns.heatmap(corr, cmap='coolwarm', annot=True) #plotando matriz
```

```python
#calculando covariância no conjunto de dados
iris.cov() 
```

```python
covar = iris.cov() #salvando covariância em um objeto
sns.heatmap(covar, cmap='coolwarm', annot=True) #plotando matriz de variância-covariância
```

```python
#gráfico com clusterização dos dados
iris_modificado = iris.drop('species', axis=1) #retirando dados de texto
sns.clustermap(iris_modificado, standard_scale=1) #criando o gráfico
```

```python
#grafico com clusters
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
```
