## Plots categóricos 

**Abra o conjunto de dados do titanic. Mostre as linhas inciais do `DataFrame`.**

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

**Utilizando gráficos de barras do `seaborn`, exiba quantos homens e mulheres sobreviveram e não sobreviveram.**

```python
sns.countplot(x='survived', data=titanic, hue='sex')
```

**Quantas pessoas sobreviveram em cada classe (coluna `class`)?**

```python
sns.barplot(data=titanic, x='class', y='survived')
```

**Verifique agora quantas pessoas exitem em cada classe (coluna `class`).** 

```python
sns.countplot(data=titanic, x='class')
```

**Construa o histograma dos dados da idade dos usuários.** <br>
**Obs: você deverá fazer um `dropna` na coluna de idade, pois em algumas linhas não temos os valores de idade.** 

```python
sns.distplot(titanic['age'].dropna(), kde=False)
```

**Por fim, faça dois histogramas para as idades, considerando separadamente pessoas que sobreviveram e não sobreviveram. Para isso, crie dois `dataframes` novos, um com as pessoas que sobreviveram outro com as pessoas que não sobreviveram.**

```python
sobreviventes = titanic[titanic['survived']==1]
naoSobreviventes = titanic[titanic['survived']==0]
plt.figure(figsize=[20,7])
plt.subplot(2,2,1)
plt.title('Sobreviventes')
sns.distplot(sobreviventes['age'].dropna())

plt.subplot(2,2,4)
plt.title('Não Sobreviventes')
sns.distplot(naoSobreviventes['age'].dropna())
```
