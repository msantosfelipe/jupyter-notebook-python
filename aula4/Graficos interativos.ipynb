**Importando modulos**

```python
import plotly.graph_objects as go
import pandas as pd
```

**Um grafico interativo simples**

```python
barras = go.Bar(y=[2,3,1]) #dados
layout = dict(title='Meu grafico',
             xaxis=dict(title='Categorias'),
             yaxis=dict(title='Quantidade'),
             width=600) # criando o layout
figura = go.Figure(data=barras, layout=layout) # criando a figura
figura.show()
```

## Usando o plotly.express

**Importanto o módulo**

```python
import plotly.express as px
```

```python
px.bar(x='Sex', color='Sex', data_frame=dados)
```

```python
px.bar(x='Sex', y='Survived', data_frame=dados)
```

**Scatter plot**

```python
iris = px.data.iris()
```

```python
iris.head()
```

```python
fig = px.scatter(x='sepal_width',
                y='sepal_length',
                data_frame=iris)
fig.show()
```

```python
fig = px.scatter(x='sepal_width',
                y='sepal_length',
                 color='species',
                data_frame=iris)
fig.show()
```

```python
fig = px.scatter(x='sepal_width',
                y='sepal_length',
                 color='species',
                 size='petal_length',
                data_frame=iris)
fig.show()
```

*Com mais informação nas margens:*

```python
fig = px.scatter(x='sepal_width',
                y='sepal_length',
                 color='species',
                 size='petal_length',
                 marginal_x='histogram',
                 marginal_y='rug',
                data_frame=iris)
fig.show()
```

```python
fig = px.scatter(x='sepal_width',
                y='sepal_length',
                 color='species',
                 size='petal_length',
                 marginal_x='violin',
                 marginal_y='box',
                data_frame=iris)
fig.show()
```

```python
fig = px.scatter(x='sepal_width',
                y='sepal_length',
                 color='species',
                 size='petal_length',
                 marginal_x='violin',
                 marginal_y='box',
                 trendline='ols', #outra opcao: lowess
                data_frame=iris)
fig.show()
```

*Um painel descritivo para os gráficos*

```python
tips = px.data.tips()
tips.head()
```

```python
fig = px.scatter(x='total_bill',
                y='tip',
                facet_row='time',
                 facet_col='day',
                 color='smoker',
                 category_orders={'day':['Thur','Fri','Sat','Sun'],
                                  'time':['Lunch','Dinner']},
                data_frame=tips)
fig.show()
```

*Explorando relações nos dados*

```python
fig = px.parallel_categories(data_frame=tips,
                            color='size',
                            color_continuous_scale=px.colors.sequential.Inferno)
fig.show()
```

**Gráficos de linhas**

```python
gapminder = px.data.gapminder()
gapminder.head()
```

```python
fig = px.line(x='year',
             y='lifeExp',
              color='continent',
             data_frame=gapminder)
fig.show()
```

**Mapas**

```python
fig = px.scatter_geo(locations='iso_alpha',
                    color='continent',
                    hover_name='country',
                    size='pop',
                    animation_frame='year',
                    projection='natural earth',
                    data_frame=gapminder)
fig.show()
```

```python
fig = px.line_geo(locations='iso_alpha',
                 color='continent',
                 projection='orthographic',
                 data_frame=gapminder)
fig.show()
```

```python
fig = px.choropleth(gapminder,
                    color='lifeExp',
                   locations='iso_alpha',
                   hover_name='country',
                   animation_frame='year',
                   range_color=[20,80])
fig.show()
```
