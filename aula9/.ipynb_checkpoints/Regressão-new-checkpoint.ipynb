# <center>Regressão</center>

## Regressão linear simples

![grafico](regressao.png)

Assume que:

- uma relação linear enre variável independente e dependente

- assume que o resíduo se distribui de forma "normal" em torno da reta

- assume que as variáveis envolvidas são numéricas e contínuas

$$\hat{y}_{i} = \beta_{1} + \beta_{2} x_{i} + \epsilon_{i}$$

Onde: $\beta$ são parâmetros da equação e $\epsilon$ é o erro aleatório associado à variável. O objetivo da regressão é estimar valores para esses parâmetros que resultam na reta que melhor "se ajusta" aos dados. Ou seja:

$$\hat{Y} = \beta_{1} + \beta_{2} X + \epsilon$$

$$\hat{\epsilon} = \hat{Y} - (\beta_{1} + \beta_{2} X)$$

***

**Importando módulos**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Criando dados de exemplo**

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000,
                      n_features=1,
                      noise=50)
```

```python
plt.scatter(X, y)
```

**Criando o modelo de regressão linear**

```python
#importando móduo de sklearn para regressão linear
from sklearn.linear_model import LinearRegression
```

```python
#criando o modelo
modelo = LinearRegression()
```

```python
#ajustando model aos dados
modelo.fit(X, y)
```

```python
#checando valor do itercepto ("beta 1", no modelo teórico)
modelo.intercept_
```

```python
#checando valor do coeficiente de inclinação da reta ("beta 2", no modelo teórico)
modelo.coef_
```

**Valores preditos pelo modelo**

```python
#realizando predições para cada ponto dos dados
y_pred = modelo.predict(X)
```

```python
plt.scatter(X, y)
plt.scatter(X, y_pred)
```

```python
#distribuição do resíduo do modelo
plt.hist(y_pred - y);
```

**Avaliando o ajuste do modelo**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_true=y, y_pred=y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse #conserva a unidade de medida da variável y
```

```python
r2_score(y_true=y, y_pred=y_pred) #coeficiente de determinação
```

**Performance preditiva**

No contexto do *machine learning*, é interessante que nossos algoritmos tenham boa performance na realização de predições. Isso requer uma boa capacidade de generalizar o que foi aprendido com um determinado conjunto de dados quando dados diferentes forem apresentados. Um problema bastante importante neste contexto é o [**overfitting**](https://en.wikipedia.org/wiki/Overfitting) (sobreajuste aos dados). 

A principal estratégia para verificar a performance preditiva dos nossos modelos é particionar o conjunto de dados em duas porções: uma para **treino** e outra para **teste** do modelo. Assim, podemos checar como ele se tenderá a se comportar em novos conjuntos de dados.

```python
from sklearn.model_selection import train_test_split
```

```python
xTreino, xTeste, yTreino, yTeste = train_test_split(X, y, train_size=0.3)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(xTreino, yTreino)
```

```python
modelo.score(xTreino, yTreino)
```

```python
modelo.score(xTeste, yTeste)
#comparar com o anterior, para checar overfitting 
#REFERENCIA: se com os dados de teste deu maior, então será overfitting
```

### Regressão linear múltipla

Frequentemente, estamos interessados em um modelo que tem mais que uma variável preditora afetando uma variável resposta. Felizmente, a regressão linear pode ser expandida para conter múltiplas variáveis resposta.

- Regressão linear simples:

$$\hat{y}_{i} = \beta_{1} + \beta_{2} x_{i} + \epsilon_{i}$$

- Regressão linear múltipla:

$$\hat{y}_{i} = \beta_{1} + \beta_{2} w_{i} + \beta_{3} x_{i} + \beta_{4} z_{i} + ... + \epsilon_{i}$$

**Criando dados de exemplo**

```python
X, y  = make_regression(n_samples=1000,
                       n_features=2,
                       noise=50)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.scatter(X.transpose()[0], y)

plt.subplot(1,2,2)
plt.scatter(X.transpose()[1], y)
```

**Criando o modelo de regressão múltipla**

```python
modelo = LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

**Valores preditos**

```python
y_pred = modelo.predict(X)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.scatter(X.transpose()[0], y) #valor observado
plt.scatter(X.transpose()[0], y_pred) #valor predito pelo modelo

plt.subplot(1,2,2)
plt.scatter(X.transpose()[1], y)
plt.scatter(X.transpose()[1], y_pred)
```

**Avaliando o ajuste do modelo**

```python
mse = mean_squared_error(y_true=y, y_pred=y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2_score(y_true=y, y_pred=y_pred)
```

### Regressão polinomial

Em alguns casos, a relação entre as variáveis pode ser não linear. Nesses casos a regressão polinomial pode representar o modelo mais adequado.

- Regressão linear simples:

$$\hat{y}_{i} = \beta_{1} + \beta_{2} x_{i} + \epsilon_{i}$$

- Regressão linear múltipla:

$$\hat{y}_{i} = \beta_{1} + \beta_{2} x_{i} + \beta_{3} {x_{i}}^{2} + \epsilon_{i}$$

```python
np.random.seed(42) #padronizando a producao de números aleatorios entre diferentes usuarios
```

```python
#criando alguns dados
x = 2 - 3*np.random.normal(0, 1, 20)
y = x - 2*(x**2) + 0.5*(x**3) + np.random.normal(-3,3,20)
```

```python
x = x[:, np.newaxis]
y = y[:, np.newaxis]
```

```python
plt.scatter(x,y)
```

**Criando o modelo**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_features = PolynomialFeatures(degree=3)
```

```python
x_poly = polynomial_features.fit_transform(x)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x_poly, y)
```

```python
y_pred = modelo.predict(x_poly)
```

```python
np.sqrt(mean_squared_error(y, y_pred))
```

```python
r2_score(y, y_pred)
```
