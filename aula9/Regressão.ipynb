### Regressão linear simples

Importando módulos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Criado dados de exemplo**

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000,
                      n_features=1,
                      noise=50)
```

```python
plt.scatter(X, y)
```

Criando modelo de regressão linear

```python
from sklearn.linear_model import LinearRegression
```

```python
#criando o modelo
modelo = LinearRegression()
```

```python
modelo
```

```python
# ajustando modelo aos dados
modelo.fit(X, y)
```

```python
#checando valor do intercepto ("beta 1" no modelo teórico)
modelo.intercept_
```

```python
#checando valor do coeficiente de inclinação da reta ("beta 2" no modelo teórico)
modelo.coef_
```

**Valores preditos pelo modelo**

```python
#realizando predições para cada ponto dos dados
y_pred = modelo.predict(X)
```

```python
plt.scatter(X, y)
plt.scatter(X, y_pred)
```

```python
plt.hist(y_pred - y)
```

**Avaliar o ajuste do modelo**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_true=y, y_pred=y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse #conserva a variável de medida da variável y
```

```python
r2_score(y_true=y, y_pred=y_pred)
```

**Performace preditiva**

```python
from sklearn.model_selection import train_test_split
```

```python
xTreino, xTeste, yTreino, yTeste = train_test_split(X, y, train_size=0.3)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(xTreino, yTreino)
```

```python
modelo.score(xTreino, yTreino)
```

```python
modelo.score(xTeste, yTeste) 
#comparar com anterior para checar overfitting
#REFERENCIA: se com os dados de teste deu maior, então será overfitting
```

### Regressão polimonial

```python
np.random.seed(42) # *padronizando a produção de numeros aleatorios entre diferentes usuarios para a aula
```

```python
#criando alguns dados
x = 2 - 3*np.random.normal(0,1,20)
y = x - 2*(x**2) + 0.5*(x**3) + np.random.normal(-3,3,20)
```

```python
x = x[:, np.newaxis]
y = y[:, np.newaxis]
```

```python
plt.scatter(x,y)
```

**Criando o modelo**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polymonial_features = PolynomialFeatures(degree=3)
```

```python
x_poly = polymonial_features.fit_transform(x)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x_poly, y)
```

```python
y_pred = modelo.predict(x_poly)
```

```python
np.sqrt(mean_squared_error(y, y_pred))
```

```python
r2_score(y, y_pred)
```

### Atividade

```python
bikes = pd.read_csv('capital_share_bike.csv')
```

```python
bikes.head()
```

```python
bikes = bikes.set_index(pd.to_datetime(bikes['datetime']))
```

```python
bikes['count'].plot()
```

```python
bikes['2011']['count'].plot()
```

```python
bikes['2011-3']['count'].plot()
```

```python
bikes[pd.isna(bikes)]
```

```python
#uso de bikes (count) em 2011 pelo tempo (weather)
#bikes['2011']['count']
sns.boxplot(x='weather', y='count', hue='season', data=bikes['2011'])
plt.legend(bbox_to_anchor=(1,1))
```

**Agrupando dados por dia**

```python
diaBikes = bikes.groupby(pd.Grouper(freq='d')).mean().dropna()
```

```python
diaBikes.head()
```

```python
y = diaBikes['count']
x = diaBikes.drop(['casual','registered','count'], axis=1)
```

```python
sns.pairplot(x)
```

```python
# verificando correlação de variáveis preditoras
x.corr().style.background_gradient(cmap='coolwarm')
```

```python
x.drop('atemp', axis=1)
```

**Implementando modelo**

```python
modelo = LinearRegression()
```

```python
modelo.fit(x, y)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
y_pred = modelo.predict(x)
```

```python
y_pred = pd.Series(y_pred, index=y.index)
```

```python
plt.plot(y)
plt.plot(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_true=y, y_pred=y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

**Modelo autoregressivo**

```python
from statsmodels.tsa.ar_model import AR
```

```python
modelo_AR = AR(y)
```

```python
modelo_AR = modelo_AR.fit()
```

```python
#melhor lag tempoal para o nosso modelo
modelo_AR.k_ar
```

```python
modelo_AR.params
```

```python
y_predict = modelo_AR.predict()
```

```python
plt.plot(y)
plt.plot(y_predict)
```

```python
mse = mean_squared_error(y[y_predict.index.min():], y_predict)
rmse = np.sqrt(mse)
rmse
```

```python
r2_score(y[y_predict.index.min():], y_predict)
```
